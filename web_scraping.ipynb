#1.4
###Escape Sequences in Python

But computer do not understand human readable format, it understands data having **\n** in it, **\t** in it and other such shortcuts listed below with meaning.

\n: Newline character - inserts a newline.

\t: Tab character - inserts a horizontal tab.

\\: Backslash character - inserts a backslash.

\\': Single quote character - inserts a single quote.

\\": Double quote character - inserts a double quote.

\r: Carriage return - moves the cursor to the beginning of the line.

\b: Backspace - moves the cursor one position back.

\f: Form feed - inserts a page break.

\v: Vertical tab - inserts a vertical tab.

\a: Bell character - produces an audible beep.

\uXXXX: Unicode escape sequence for a character in the Basic Multilingual Plane, where XXXX represents the character's Unicode code point in hexadecimal.

\UXXXXXXXX: Unicode escape sequence for a character outside the Basic Multilingual Plane, where XXXXXXXX represents the character's Unicode code point in hexadecimal.

```python
print("This is a\ttabbed\tline.")
# Output: This is a    tabbed  line.

print("This is a new\nline.")
# Output:
# This is a
# line.

print("This is a backslash: \\")
# Output: This is a backslash: \

print("This is a unicode: \u1234")
# Output: This is a unicode: 



# try Ruppee () symbol unicode "20B9"

# try other escape sequences by yourself by searching on google
```

#2.2a - Single Device Name
Give a start word and a end word.

Find that start and end words in the above text.

Filter in data written between start and end words.


```python
start_word = "Add to Compare"
end_word = "("
start_index = text.find(start_word)
end_index = text.find(end_word, start_index + len(start_word))
name = ''
if start_index != -1 and end_index != -1:
    name = text[start_index + len(start_word):end_index].strip()

print(name)
```

#2.2b - Single Device RAM and ROM

```python
start_word = "Reviews"
end_word = "ROM"
start_index = text.find(start_word)
end_index = text.find(end_word, start_index + len(start_word))
RAM_ROM = ''
if start_index != -1 and end_index != -1:
    RAM_ROM = text[start_index + len(start_word):end_index].strip()

print(RAM_ROM)
```

#2.2c - Single Device Rating and Sold

```python
start_word = "GB)"
end_word = "Ratings"
start_index = text.find(start_word)
end_index = text.find(end_word, start_index + len(start_word))
Rating_Sold = ''
if start_index != -1 and end_index != -1:
    Rating_Sold = text[start_index + len(start_word):end_index].strip()

print(Rating_Sold)
```

#2.3 Mobile Names

Add a for loop to filter all mobile names.

```python
import re

start_word = "Add to Compare"
end_word = "("

escaped_start_word = re.escape(start_word)
escaped_end_word = re.escape(end_word)

pattern = f"{escaped_start_word}(.*?){escaped_end_word}"
matches = re.finditer(pattern, text)
name = []

for match in matches:
    project = match.group(1).strip()
    name.append(project)

print(name)
```

```python
name
```

#2.4 RAM and ROM

```python
import re

start_word = "Reviews"
end_word = "ROM"

escaped_start_word = re.escape(start_word)
escaped_end_word = re.escape(end_word)

pattern = f"{escaped_start_word}(.*?){escaped_end_word}"
matches = re.finditer(pattern, text)
RAM_ROM = []

for match in matches:
    project = match.group(1).strip()
    RAM_ROM.append(project)

print(RAM_ROM)
```

```python
RAM_ROM
```

#2.4a

Sorting Data by removing last 5 elements.

```python
new_RAM_ROM = RAM_ROM[:-5]  # Remove the last 5 elements

print(new_RAM_ROM)
```

```python
new_RAM_ROM
```

#2.4b
Add **"0 GB RAM | "** in which RAM is not given.

```python
modified_RAM_ROM = [element if '|' in element else '0 GB RAM | ' + element for element in new_RAM_ROM]

print(modified_RAM_ROM)
```

```python
modified_RAM_ROM
```

#2.4c
Split RAM and ROM in 2 differnt sets.

```python

RAM = [element.split(" | ")[0] for element in modified_RAM_ROM]
ROM = [element.split(" | ")[1] for element in modified_RAM_ROM]

print(RAM)
print(ROM)
```

#2.4d
Keep only numerics in the set.

```python
RAM_digit = [re.sub(r'\D', '', element) for element in RAM]

print(RAM_digit)
```

```python
ROM_digit = [re.sub(r'\D', '', element) for element in ROM]

print(ROM_digit)
```

#2.5 Rating and Sold Quantity

```python
import re

start_word = "GB)"
end_word = "Rating"

escaped_start_word = re.escape(start_word)
escaped_end_word = re.escape(end_word)

pattern = f"{escaped_start_word}(.*?){escaped_end_word}"
matches = re.finditer(pattern, text)
Rating_and_Sold = []

for match in matches:
    project = match.group(1).strip()
    Rating_and_Sold.append(project)

print(Rating_and_Sold)
```

```python
Rating_and_Sold
```

#2.5a

Spliting into 2 sets based on selecting first 3 elements.

```python
# Splitting into two sets
Rating = [element[:3] for element in Rating_and_Sold]
Sold = [element[3:] for element in Rating_and_Sold]

print(Rating)
print(Sold)
```

```python
Rating
```

```python
Sold
```

#2.5b
Removing all commas

```python
# Remove commas from the elements
Sold_Data = [element.replace(',', '') for element in Sold]

print(Sold_Data)
```

```python
Sold_Data
```

#2.6 Exporting the final CSV

```python
import csv
```

```python
data = zip(name ,RAM_digit, ROM_digit, Sold_Data, Rating)

filename = 'merged_data.csv'

with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Name', 'RAM', 'ROM', 'Sold Quantity', 'Rating'])  # Write header row
    writer.writerows(data)

print(f"Data has been written to {filename}.")
```

```python
prefixed_links = [f"https://www.flipkart.com/{link}" for link in links]

for link in prefixed_links:
    print(link)
```

```python
import csv
filename = "links.csv"

with open(filename, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Links"])  # Write header row
    writer.writerows([[link] for link in prefixed_links])  # Write links row by row

print(f"CSV file '{filename}' created successfully.")
```
